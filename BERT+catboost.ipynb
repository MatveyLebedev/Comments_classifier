```python
import pandas as pd
from sklearn.metrics import accuracy_score, balanced_accuracy_score
from catboost import CatBoostClassifier
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
df = pd.read_excel('Data_scientist_NLP_данные.xlsx')
df = df.dropna()
df
```

```python
vec = np.load('text_vectors.npy')
vec.shape
```

```python
dfv = pd.DataFrame(vec)
dfv[['Категория', 'Тема', 'Муниципалитет']] = df[['Категория', 'Тема', 'Муниципалитет']]
dfv = dfv.dropna()
dfv
```

```python
le1 = LabelEncoder()
le2 = LabelEncoder()
ley = LabelEncoder()

dfv['Категория'] = ley.fit_transform(dfv['Категория'])
dfv['Тема'] = le1.fit_transform(dfv['Тема'])
dfv['Муниципалитет'] = le2.fit_transform(dfv['Муниципалитет'])
```

```python
X = dfv.drop(columns=dfv['Категория'])
y = dfv['Категория']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
clf = CatBoostClassifier().fit(X_train, y_train, eval_set=(X_test, y_test))
```

```python
preds = clf.predict(X_test)
print(balanced_accuracy_score(preds, y_test))
print(classification_report(preds, y_test, target_names=['Вопрос', 'Жалоба']))
```


Вывод:
Catboost отлично справляется с классификацией эмбеденгов полученных с помощю BERT.
